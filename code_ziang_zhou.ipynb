```python
import pandas as pd
data = pd.read_csv(file_path_data)
sensor = pd.read_csv(file_path_sensor)
```

```python
id_loc= {}
for i in range(sensor.shape[0]):
  id = sensor.iloc[i,0]
  loc = sensor.iloc[i,-1]
  id_loc[id]=loc
```

After checking, "location" and "Sensor_ID" are corresponded to each other. Thus, I make a mapping for them.

```python
data["location"]=data["Sensor_ID"].map(id_loc)
```

# Q1: Top 10 (most pedestrians) locations by day

```python
#Using groupy and agg to sum up Hourly_Counts
ans_q1_df = data.groupby(["Year","Month","Mdate","location"]).agg({"Hourly_Counts":"sum"}).sort_values("Hourly_Counts", ascending = False)[0:11]
ans_q1_df
```

# Q2: Top 10 (most pedestrians) locations by month

```python
#Using groupy and agg to sum up Hourly_Counts
ans_q2_df = data.groupby(["Year","Month","location"]).agg({"Hourly_Counts":"sum"}).sort_values("Hourly_Counts", ascending = False)[0:11]
ans_q2_df
```

# Q3: Which location has shown most decline due to lockdowns in last 2 years

```python
#Using groupy and agg to sum up Hourly_Counts
ans_year_df = data.groupby(["Year","location"]).agg({"Hourly_Counts":"sum"}).sort_values("Year", ascending = False).reset_index()
ans_year_df
```

```python
# Select 2022 and 2020's data from the whole dataset
two_year_df = ans_year_df.loc[(ans_year_df["Year"] == 2020) | (ans_year_df["Year"] == 2022)].sort_values("location", ascending = False).reset_index()
two_year_df = two_year_df.rename(columns={"index":"index_1"})
two_year_df
```

```python
# Delect sensors which only have data for one year
tmp_list = []
for i in two_year_df.index:
  if i == 0:
    if two_year_df.iloc[i].location != two_year_df.iloc[i+1].location:
      tmp_list.append(two_year_df.iloc[i].index_1)
  elif i == two_year_df.index[-1]:
    if two_year_df.iloc[i].location != two_year_df.iloc[i-1].location:
      tmp_list.append(two_year_df.iloc[i].index_1)
  else:
    if two_year_df.iloc[i].location != two_year_df.iloc[i+1].location and two_year_df.iloc[i].location != two_year_df.iloc[i-1].location:
      tmp_list.append(two_year_df.iloc[i].index_1)
tmp_list
```

```python
for i in tmp_list:
  two_year_df = two_year_df.drop(two_year_df[two_year_df.index_1 == i].index)
two_year_df
```

```python
# this function calculates the precent change of Hourly_Counts
def fun1(col):
  pre = col[col.Year==2020].Hourly_Counts
  cur = col[col.Year==2022].Hourly_Counts 
  for i,j in zip(pre,cur):
    return i/j-1
```

```python
ans_q3_df = two_year_df.groupby("location").apply(fun1).to_frame().reset_index().rename(columns={0:"rate"}).sort_values("rate", ascending = True).head(10)
ans_q3_df
```

```python
# Delect sensors which only have data for one year
tmp_list = []
for i in one_year_df.index:
  if i == 0:
    if one_year_df.iloc[i].location != one_year_df.iloc[i+1].location:
      tmp_list.append(one_year_df.iloc[i].index_1)
  elif i == one_year_df.index[-1]:
    if one_year_df.iloc[i].location != one_year_df.iloc[i-1].location:
      tmp_list.append(one_year_df.iloc[i].index_1)
  else:
    if one_year_df.iloc[i].location != one_year_df.iloc[i+1].location and one_year_df.iloc[i].location != one_year_df.iloc[i-1].location:
      tmp_list.append(one_year_df.iloc[i].index_1)
tmp_list
```

```python
for i in tmp_list:
  one_year_df = one_year_df.drop(one_year_df[one_year_df.index_1 == i].index)
ans_q4_df = two_year_df.groupby("location").apply(fun1).to_frame().reset_index().rename(columns={0:"rate"}).sort_values("rate", ascending = False).head(1)
ans_q4_df
```
